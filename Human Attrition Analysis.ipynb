# EDA and Data Visualisation

```python
Attrition_Count=hrdata_df['Attrition'].value_counts()
print(Attrition_Count)
Attrition_Percent=hrdata_df['Attrition'].value_counts(normalize=True)
print(Attrition_Percent)
print("*"*125)
fig,ax=plt.subplots(1,1,figsize=(10,5))
hrdata_df['Attrition'].value_counts().plot(kind='pie',autopct="%.2f", shadow=True,startangle=90,labels=Attrition_Count.index, ax=ax)
plt.title("Attrition_Count")
#plt.legend(labels=labels, loc='upper right')
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
sns.countplot(x="Attrition", data=hrdata_df, ax=ax)
plt.title("Attrition_Count")
plt.ylabel("Amount_of_Attrition")

for bars in ax.containers:
    ax.bar_label(bars)
```
