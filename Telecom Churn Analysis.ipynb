```python
Churn_Count=telecom_df['Churn'].value_counts()
print(Churn_Count)
print("="*125)
Churn_Percent=telecom_df['Churn'].value_counts(normalize=True)
print(Churn_Percent)
print("*"*125)

fig,ax=plt.subplots(1,1,figsize=(10,5))
sns.countplot(y="Churn", data=telecom_df, ax=ax, orient="h")
plt.title("Amount Of Churn Count")
plt.ylabel("Churn_Count")
```

```python
Top_10_States=((telecom_df.groupby(['State'])['Churn'].mean())*100).head(10).sort_values(ascending=False).plot(kind='bar')
print(Top_10_States)
plt.title("State With most Churn Percent")
plt.ylabel("Amount_Of_Churn")
plt.xlabel("State")
#sns.countplot(y="State", order=telecom_df['State'].value_counts().index[0:10], data=telecom_df)
```

```python
State_data=pd.crosstab(telecom_df['State'], telecom_df['Churn'])
State_data['Percentage_Churn']=State_data.apply(lambda x:x[1]*100/x[0]*x[1], axis=1)
print(State_data)
```

```python
State_Wise_Churn=telecom_df.groupby(['State'])['Churn'].value_counts().unstack()
print(State_Wise_Churn)
State_Wise_Churn.plot(kind='bar', stacked='False', figsize=(10,5))
```

```python
NO_International_Plan=telecom_df[telecom_df['International plan']=="No"]
YES_International_Plan=telecom_df[telecom_df['International plan']=="Yes"]
```

```python
NO_International_Plan_Churn_Count=NO_International_Plan['Churn'].value_counts()
YES_International_Plan_Churn_Count=YES_International_Plan['Churn'].value_counts()
print("NO_International_Plan_Churn_Count")
print(NO_International_Plan_Churn_Count)
print("-"*125)
print("YES_International_Plan_Churn_Count")
print(YES_International_Plan_Churn_Count)
print("="*125)
NO_International_Plan_Churn_Percent=NO_International_Plan['Churn'].value_counts(normalize=True)
YES_International_Plan_Churn_Percent=YES_International_Plan['Churn'].value_counts(normalize=True)
print("NO_International_Plan_Churn_Percent")
print(NO_International_Plan_Churn_Percent)
print("-"*125)
print("YES_International_Plan_Churn_Percent")
print(YES_International_Plan_Churn_Percent)
print("*"*125)

fig=plt.figure()
ax0=fig.add_subplot(121)
ax1=fig.add_subplot(122)
NO_International_Plan_Churn_Count.plot(kind='pie',autopct="%.2f",shadow=True, labels=NO_International_Plan_Churn_Count.index, figsize=(10,5), ax=ax0)
ax0.set_title("NO_International_Plan_Churn_Count")
#ax0.set_ylabel("NO_International_Plan_Churn")
#ax0.set_xlabel("Churn")
YES_International_Plan_Churn_Count.plot(kind='pie', autopct="%.2f",shadow=True, labels=YES_International_Plan_Churn_Count.index, figsize=(10,5), ax=ax1)
ax1.set_title("YES_International_Plan_Churn_Count")
#ax1.set_ylabel("YES_International_Plan_Churn")
#ax1.set_xlabel("Churn")
```

```python
Total_Day_Charge_Churn=telecom_df.groupby(['Churn'])['Total day charge'].sum()
print(Total_Day_Charge_Churn)
fig=plt.figure()
ax0=fig.add_subplot(121)
ax1=fig.add_subplot(122)
Total_Day_Charge_Churn.plot(kind='pie', autopct="%.2f", shadow=True, labels=Total_Day_Charge_Churn.index, ax=ax0)
Total_Day_Charge_Churn.plot(kind='bar',  figsize=(10,5), color='green', ax=ax1)
plt.tight_layout(pad=2.00)

for bars in ax1.containers:
    ax1.bar_label(bars)
```
